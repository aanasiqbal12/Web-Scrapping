```python
import requests
weblink = "http://books.toscrape.com/"

page = requests.get(weblink).text
soup = BeautifulSoup(page,"html.parser")
#print(soup)
artists = soup.find_all('a')
for artist in artists:
    names = artist.contents[0]
    fulllink = artist.get("href")
    print(names)
    print(fulllink)
```
